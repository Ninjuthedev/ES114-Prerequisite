```python
# Problem 1
list1 = []
even_sum = 0
odd_product = 1
mean = 0

for i in range(1,21):
    list1.append(i)
length_list = len(list1)
for j in list1:
    if j%2 == 0:
        even_sum+=j
        mean += j
    else:
        odd_product *= j
        mean += j
print(f'Sum of even numbers: {even_sum} ')
print(f"Product of Odd Numbers: {odd_product}")
mean /= length_list
print(f"Mean: {mean}")
list1.sort()
if length_list % 2 !=0:
    median = list1[length_list//2]
else:
    median = (list1[(length_list//2)-1] + list1[(length_list//2)])/2
print(f"Median: {median}")
tempvar = 0
for k in list1:
    tempvar += ((k-mean)**2)
standard_deviation = (tempvar/length_list)**0.5
print(f"Standard Deviation: {standard_deviation}")
```

```python
# Problem 21
list1 = [7, 24, 12, 35, 48, 3, 19, 10, 27, 16, 8, 44, 30, 6, 21, 18, 50, 11, 40, 23, 7]
list1.sort()
total = 0
for i in list1:
    total += i
mean = total / len(list1)
print(f"Mean: {mean}")
length_list = len(list1)
if length_list % 2 !=0:
    median = list1[length_list//2]
else:
    median = (list1[(length_list//2)-1] + list1[(length_list//2)])/2
print(f"Median: {median}")
tempvar = 0
for k in list1:
    tempvar += ((k-mean)**2)
standard_deviation = (tempvar/length_list)**0.5
print(f"Standard Deviation: {standard_deviation}")
frequency = {}
for num in list1:
    if num in list(frequency.keys()):
        frequency[num] += 1
    else:
        frequency[num] = 1
print(f"Mode: {max(list(frequency.values()))}")
plt.boxplot(list1)
plt.title("Box Plot")
plt.xlabel("Data")
plt.ylabel("Values")
plt.show()
```

```python
# Problem 22
def normalize(x):
    mylist = []
    minimumval = sorted(x)[0]
    maximumval = sorted(x)[-1]
    for i in x:
        normalizedval = ((i-minimumval)/(maximumval - minimumval))
        mylist.append(normalizedval)
    return mylist
n = eval(input())
print(f"Original Data: {n}")
print(f"Normalized data: {normalize(n)}")
plt.plot(n,normalize(n))
plt.xlabel("Original Data")
plt.ylabel("Normalized Data")
plt.title("Line Plot of Normalized Data")
plt.show()
```
